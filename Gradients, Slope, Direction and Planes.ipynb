## Gradient Plane, Slope and Direction and Tangent Plane


Let's define a parabolic function:

$$
z = f( x, y) = x ^ 2 + y ^ 2
$$

## 1. Parabola Plot

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ---------------- Paraboloid Function ----------------
def f(x, y):
    return x**2 + y**2

# ---------------- Paraboloid Points ----------------
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# ---------------- 3D Figure ----------------
fig_3d = go.Figure()

# Paraboloid surface
fig_3d.add_trace(go.Surface(x=X, y=Y, z=Z, colorscale="Viridis", opacity=0.7, name="Surface"))

fig_3d.update_layout(
    title="3D Paraboloid Surface",
    scene=dict(
        xaxis_title="X-axis",
        yaxis_title="Y-axis",
        zaxis_title="Z-axis",
        aspectratio=dict(x=1.5, y=1.5, z=1,)
    ),
    width=900,
    height=600
)
```

At any point $(x_0, y_0)$ on this surface, the **gradient vector** is:

$$
\nabla f(x, y) =
\left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
=
\left( 2 x, 2 y \right)
$$

Thus at any point $(x_0, y_0)$ gradient is:

$$
\nabla f(x_0, y_0)
=
\left( 2 x_0, 2 y_0 \right)
$$


Let's construct a tangent line at the point:

$$
(x_0, y_0) = (1, 1)
$$

Which yields: $z_0=2$

## 2. Calculating Slope:

$$
\nabla f(x_0, y_0)
=
\left( 2 x_0, 2 y_0 \right)
=
\left( 2, 2 \right)
$$



```python
# ---------------- Chosen Point ----------------
x0, y0 = 1.0, 1.0
z0 = f(x0, y0)

# ---------------- Tangent Slopes ----------------
dz_dx = 2 * x0
dz_dy = 2 * y0
```

## 3. Tangent Lines And Direction

- Length Of Tangent Line In PLot = 1.5
- Number of Points in Tangent Line = 50
- Tangent Arrow Cone Size = 0.5 for Tangent Direction

```python
L = 1.5  # length of tangent lines
arrow_size = 0.5 # size of the arrows cone
```

### 3.1. Tangent Accross X

Here, we keep values of y constant i.e. $y=y_0$

We Know,

$$
\frac{\partial z}{\partial x} = \frac{z-z_0}{x-x_0}
$$

Which Yields:

$$
z = z_0 + \frac{\partial z} {\partial x} (x - x_0)
$$

So, we need to figue out values of z for a given x keeping y constant

```python
# ---------------- Tangent Lines in 3D ----------------
x_tangent_x = np.linspace(x0 - L, x0 + L, 50)
y_tangent_x = np.full_like(x_tangent_x, y0) # constant values of y = y0
z_tangent_x = z0 + dz_dx * (x_tangent_x - x0)

# Tangent lines
fig_3d.add_trace(go.Scatter3d(x=x_tangent_x, y=y_tangent_x, z=z_tangent_x,
                              mode="lines", line=dict(color="blue", width=6), name="Tangent X"))

# -------------- Tangent arrows: as a cone -------------------------
fig_3d.add_trace(go.Cone(x=[x_tangent_x[-1]], y=[y_tangent_x[-1]], z=[z_tangent_x[-1]],
                         u=[1], v=[0], w=[dz_dx], colorscale=[[0,"blue"],[1,"blue"]],
                         sizemode="absolute", sizeref=arrow_size, showscale=False))

fig_3d.add_trace(go.Scatter3d(x=[x0], y=[y0], z=[z0], mode="markers",
                              marker=dict(size=6, color="red"), name="Point"))
```

### 3.2. Tangent Accross Y

Here, we keep values of y constant i.e. $x=x_0$

We Know,

$$
\frac{\partial z}{\partial y} = \frac{z-z_0}{y-y_0}
$$

Which Yields:

$$
z = z_0 + \frac{\partial z} {\partial y} (y - y_0)
$$

So, we need to figue out values of z for a given y keeping x constant

```python
# ---------------- Tangent Lines in 3D ----------------
y_tangent_y = np.linspace(y0 - L, y0 + L, 50)
x_tangent_y = np.full_like(y_tangent_y, x0)
z_tangent_y = z0 + dz_dy * (y_tangent_y - y0)

fig_3d.add_trace(go.Scatter3d(x=x_tangent_y, y=y_tangent_y, z=z_tangent_y,
                              mode="lines", line=dict(color="green", width=6), name="Tangent Y"))

# -------------- Tangent arrows: as a cone -------------------------
fig_3d.add_trace(go.Cone(x=[x_tangent_y[-1]], y=[y_tangent_y[-1]], z=[z_tangent_y[-1]],
                         u=[0], v=[1], w=[dz_dy], colorscale=[[0,"green"],[1,"green"]],
                         sizemode="absolute", sizeref=arrow_size, showscale=False))
```

## 4. Tangent Planes
- Length of Plane Size = 1.5
- Plane Points = 10

Equation Of Tangent Plane is Given By:


$$
z = z_0 + \frac {\partial z}{\partial x} (x-x_0) + \frac {\partial z}{\partial y} (y-y_0)
$$

```python
plane_size = 1.5
plane_points = 10

# x and y coordinates for the tangent plane
x_plane = np.linspace(x0 - plane_size, x0 + plane_size, plane_points)
y_plane = np.linspace(y0 - plane_size, y0 + plane_size, plane_points)
X_plane, Y_plane = np.meshgrid(x_plane, y_plane)

# z values for the tangent plane
# Using the tangent plane equation: z = z0 + dz_dx * (x - x0) + dz_dy * (y - y0)
Z_plane = z0 + dz_dx * (X_plane - x0) + dz_dy * (Y_plane - y0)

fig_3d.add_trace(go.Surface(x=X_plane, y=Y_plane, z=Z_plane,
                            colorscale=[[0, 'rgba(255,0,0,0.3)'], [1, 'rgba(255,0,0,0.3)']],
                            showscale=False, name="Tangent Plane", opacity=0.5))
```

## 5. Tangent Slices ( X vs Z and Y vs Z ) Plots

For better intuition, we also visualize **2D slices**:

1. **XZ slice at $y=y_0$**:

$$
z(x) = x^2 + y_0^2
$$

- The dashed red line is the **tangent at $x_0$**, slope $\frac{\partial z}{\partial x} = 2 x_0$.
- The black dot indicates the chosen point.


```python
fig_2d = make_subplots(rows=1, cols=1, subplot_titles=("X vs Z at y=y0", ))

# X-Z slice
x_slice = np.linspace(-3, 3, 200)
z_xslice = f(x_slice, y0)
z_tangent_x = z0 + dz_dx * (x_slice - x0)  # tangent line

# Add X-Z slice plot
fig_2d.add_trace(go.Scatter(x=x_slice, y=z_xslice, mode="lines", line=dict(color="blue"), name="X-Z Slice"), row=1, col=1)

# Add tangent line at y0
fig_2d.add_trace(go.Scatter(x=x_slice, y=z_tangent_x, mode="lines", line=dict(color="red", dash="dash"), name="Tangent X"), row=1, col=1)

# Add point at (x0, z0)
fig_2d.add_trace(go.Scatter(x=[x0], y=[z0], mode="markers", marker=dict(color="black", size=8), name="Point"), row=1, col=1)

fig_2d.update_layout(
    title="2D Slice at y = {}".format(y0),
    xaxis_title="X-axis",
    yaxis_title="Z-axis",
    width=900,
    height=600
)

fig_2d.show()
```

2. **YZ slice at $x=x_0$**:

$$
z(y) = x_0^2 + y^2
$$

- The dashed orange line is the **tangent at $y_0$**, slope $\frac{\partial z}{\partial y} = 2 y_0$.
- The black dot indicates the chosen point.

```python
fig_2d = make_subplots(rows=1, cols=1, subplot_titles=("Y vs Z at x=x0", ))

# Y-Z slice
y_slice = np.linspace(-3, 3, 200)
z_yslice = f(x0, y_slice)
z_tangent_y = z0 + dz_dy * (y_slice - y0)  # tangent line

fig_2d.add_trace(go.Scatter(x=y_slice, y=z_yslice, mode="lines", line=dict(color="green"), name="Y-Z Slice"), row=1, col=1)
fig_2d.add_trace(go.Scatter(x=y_slice, y=z_tangent_y, mode="lines", line=dict(color="orange", dash="dash"), name="Tangent Y"), row=1, col=1)
fig_2d.add_trace(go.Scatter(x=[y0], y=[z0], mode="markers", marker=dict(color="black", size=8), name="Point"), row=1, col=1)

fig_2d.update_layout(
    title="2D Slice at y = {}".format(x0),
    xaxis_title="Y-axis",
    yaxis_title="Z-axis",
    width=900,
    height=600
)

fig_2d.show()
```
